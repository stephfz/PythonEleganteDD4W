### 5. Operador ternario

Python no tiene operador ternario por defecto pero hay una alternativa.

Ej. Java


```
result = (num1>num2) ? (num1+num2):(num1-num2)
```



````python
# Sin conocer la alternativa al operador ternario 

num1 = 20
num2 = 5

result  = 0
is_greater = num1 > num2

if is_greater:
  result = num1 - num2
else:
  result = num1 + num2  

print(result)  
````

````python
# Conociendo la alternativa ;)
is_greater = num1 > num2
result = (num1 - num2) if is_greater else (num1 + num2)

print(result)
````

````python
# compactando todo
result = (num1 - num2) if (num1 > num2) else (num1 + num2)
print(result)
````

### 9. Comprensin de listas
En la medida de lo posible, crear listas en una sola linea para mejor legibilidad.
[Ref](https://ellibrodepython.com/list-comprehension-python)

````python
prime_numbers = list()

for element in range(10):
  if (element % 2): 
    prime_numbers.append(element)

print(prime_numbers)    
````

````python
# Un poco mejor, usando funciones lambda
prime_numbers_lambda = list(filter(lambda element: element % 2, range(10)))
print (prime_numbers_lambda)
````

````python
# Usando compresion
prime_numbers_comp = [element for element in range(10) if (element % 2)]
print (prime_numbers_comp)
````

### 10. `enumerate` para Indices y Elementos

````python
# Usando variables innecesarias

the_weekend = ['viernes', 'sabado', 'domingo']
index = 0
while index < len(the_weekend):
  print(the_weekend[index], index)
  index += 1
````

````python
#the_weekend = ['viernes', 'sabado', 'domingo']
for index, value in enumerate(the_weekend):
  print(value, index)
````

````python
# otra forma, aun mas compacta
print (list(enumerate(the_weekend)))
````

### 11. Comprensin de Diccionarios
A veces, por alguna motivo queremos "cruzar" listas para crear un diccionario.

````python
# Usando iteradores

headers = ['ciudad', 'densidad', 'poblacion']
data = ['Narnia', 45.6, 450]

dicc = {}
for header in headers:
  for dato in data:
    dicc[header] = dato


print(dicc)    
````

````python
# Usando zip

my_shiny_dict = {i:j for i,j in zip(headers, data)}

print(my_shiny_dict)  
````

### 12. Compresin de sets
Similar a la compresion de listas, solo que reemplazamos los `( )` por `{ }`

````python
# Sin compresion
frase = "La mar estaba serena, serena estaba la mar"
unique_elements = set()

unique_elements = set()
for word in frase:
  for char in word:
    unique_elements.add(char)

print(unique_elements)
````

````python
# Compresion  == one liner :)
unique_elements = {letra for letra in frase}
print (unique_elements)
````
